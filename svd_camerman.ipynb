# SVD for Camera Man

`Singular Value Decomposition`


---

In this notebook, we rely on singular value decomposition to show how matrix factorization reveals very important information about the original matrix.

## Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## Data

```python
img=mpimg.imread('cameraman.png')
imgplot = plt.imshow(img)
X = img / 255
```

## Singular value decomposition

```python
U, S, V = np.linalg.svd(X, full_matrices=True)
res = np.matmul(np.matmul(U, np.diag(S)), V)
plt.imshow(res)
```

```python
S = np.diag(S)
fig = plt.figure(figsize=(12,6), layout='tight')
index = 1
for n in [250, 150, 50, 30, 20, 10]:
    B=np.matmul(np.matmul(U[:,:n],S[:n,:n]),V[:n,:])
    ax = fig.add_subplot(1,6,index)
    ax.set_title(f"n={n}")
    ax.imshow(B)
    index += 1
```
